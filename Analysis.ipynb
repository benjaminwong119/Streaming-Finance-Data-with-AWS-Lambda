```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("results.csv")
```

```python
#df
```

# Which data is the most volatile?

```python
plt.figure(figsize=(20,20))
sns.lineplot(data=df,x="Date",y="Avg_Volatility",hue="Company")
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("Date",fontsize=17)
plt.ylabel("Average Volatility",fontsize=17)
plt.title("Average Volatility Trend per Company",fontsize=20)
```

Based on the line graph, Costco ($COST) is the most volatile stock.

# Does the daily highest volatility per company support the previous graph?

```python
sns.catplot(data=df, kind="bar",x="Company", y="Max_Volatility", hue="Date",height=15,legend=False)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("Company",fontsize=17)
plt.ylabel("Max Volatility",fontsize=17)
plt.title("Daily Highest Volatility per Company",fontsize=20)
plt.legend(fontsize=15)
```

This graph supports the last one as Costco ($COST) has the highest volatility for a majority of the days

# Normalization allows the comparison of quantities or objects on an appropriate scale. In this case, which company is actually the most volatile?

```python
df1 = df.copy(True)
#df1
```

```python
df1["Norm_Avg"]=(df1["Avg_Volatility"]-df1["Min_Volatility"])/(df1["Max_Volatility"]-df1["Min_Volatility"])
```

```python
df2= df1.groupby("Company")["Norm_Avg"].mean().reset_index()
#df2
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=df2,x="Company",y="Norm_Avg")
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel("Company",fontsize=15)
plt.ylabel("Normalized Average Volatility",fontsize=15)
plt.title("Normalized Average Volatility per Company",fontsize=17)
```

After normalizing the volatility, Amazon ($AMZN) is now the most volatile.
